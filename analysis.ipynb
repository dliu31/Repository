```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("diabetes.csv")
```

```python
data.describe()

# Display summary information about the fraudulent transactions
diabetes = data[(data["Outcome"] == 1)]
diabetes.describe()
```

```python
# Display summary information about the non-fraudulent transactions
non_diabetes = data[(data["Outcome"] == 0)]
non_diabetes.describe()
```

```python
# Display any rows with null values
print(diabetes.isnull().sum())
print(non_diabetes.isnull().sum())
# The dataset does not contain any null values.
```

# Viewing Distributions

```python
for x in data.columns:
    plt.hist(data[x], bins=15, color="skyblue", edgecolor="black")
    plt.title(x)
    plt.show()
```

# Exploratory Data Analysis

```python
# Check the boxplots of the selected features based on whether the person does or does not have diabetes
selected_features = ["Pregnancies","Insulin","SkinThickness","DiabetesPedigreeFunction","Age","Glucose","BloodPressure","BMI"]
for feature in selected_features:
    feature_data = data[[feature, "Outcome"]]
    with_diabetes = feature_data[(feature_data["Outcome"] == 1)].drop("Outcome", axis=1)
    without_diabetes = feature_data[(feature_data["Outcome"] == 0)].drop("Outcome", axis=1)
    feature_data = [without_diabetes[feature], with_diabetes[feature]]
    plt.boxplot(feature_data, positions=[0,1])
    plt.ylabel(feature)
    plt.show()
```

All factors except skin thickness seem to have an impact on whether someone is diagnosed with diabetes or not since the boxplots are all shifted slightly higher. The ones that are more likely to have an impact seem to be pregnancies, glucose, and insulin.
