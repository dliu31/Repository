```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("creditcard.csv")
```

```python
data.describe()

# Display summary information about the fraudulent transactions
fraudulent_transactions = data[(data["Class"] == 1)]
fraudulent_transactions.describe()
```

```python
# Display summary information about the non-fraudulent transactions
non_fraudulent_transactions = data[(data["Class"] == 0)]
non_fraudulent_transactions.describe()
```

```python
# Display any rows with null values
print(fraudulent_transactions.isnull().sum())
print(non_fraudulent_transactions.isnull().sum())
# The dataset does not contain any null values.
```
