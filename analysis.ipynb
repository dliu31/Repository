```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("diabetes.csv")
```

```python
data.describe()

# Display summary information about the fraudulent transactions
diabetes = data[(data["Outcome"] == 1)]
diabetes.describe()
```

```python
# Display summary information about the non-fraudulent transactions
non_diabetes = data[(data["Outcome"] == 0)]
non_diabetes.describe()
```

```python
# Display any rows with null values
print(diabetes.isnull().sum())
print(non_diabetes.isnull().sum())
# The dataset does not contain any null values.
```

# Viewing Distributions

```python
for x in data.columns:
    plt.hist(data[x], bins=15, color="skyblue", edgecolor="black")
    plt.title(x)
    plt.show()
```
